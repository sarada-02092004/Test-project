```python
import pandas as pd

# Reading the file
df = pd.read_csv('/content/homeprices (1).csv')
# Draw the Linear Graph
import matplotlib.pyplot as plt
# initializing the data
x = df.area
y = df.price
# plotting the data
plt.plot(x, y)
plt.show()
```

```python
# Syntax:matplotlib.pyplot.title(label, fontdict=None, loc=’center’, pad=None, **kwargs)

import matplotlib.pyplot as plt

# initializing the data
x = df.area
y = df.price

# plotting the data
plt.plot(x, y)

# Adding title to the plot
plt.title("Linear graph")

plt.show()
```

```python
# Example Adding X label and Y label

import matplotlib.pyplot as plt


# initializing the data
x = df.area
y = df.price

# plotting the data
plt.plot(x, y)

# Adding title to the plot
plt.title("Linear graph", fontsize=25, color="green")

# Adding label on the y-axis
plt.ylabel('Y-Axis')

# Adding label on the x-axis
plt.xlabel('X-Axis')

plt.show()
```

```python
# Histogram
import matplotlib.pyplot as plt

# initializing the data
x = df.area


# plotting the data
plt.hist(x)

# Adding title to the plot
plt.title("Area vs Price")

# Adding label on the y-axis
plt.ylabel('Price')

# Adding label on the x-axis
plt.xlabel('Area')

plt.show()
```

```python
# Draw Bar Chart
import matplotlib.pyplot as plt

# initializing the data
x = df.area
y = df.price

# plotting the data
plt.bar(x, y)
# Adding title to the plot
plt.title("Area Vs Price")

# Adding label on the y-axis
plt.ylabel('Price')

# Adding label on the x-axis
plt.xlabel('Area')
plt.show()
```

```python
# Pie Chart
import matplotlib.pyplot as plt

# initializing the data
area = df.area
data = df.price


# plotting the data
plt.pie(data, labels=area)

# Adding title to the plot
plt.title("area data")

plt.show()
```
